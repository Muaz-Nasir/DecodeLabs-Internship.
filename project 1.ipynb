```python
import pandas as pd


df = pd.read_excel("/content/Dataset for Data Analytics.xlsx")

df["CouponCode"] = df["CouponCode"].fillna("NO_COUPON")
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

invalid_dates = df["Date"].isna().sum()
df = df.drop_duplicates(subset=["OrderID"], keep="first")

print("-" * 40)
print("DATA CLEANING VERIFICATION REPORT")
print("-" * 40)
print(f"Duplicate Order IDs Remaining : {df.duplicated(subset=['OrderID']).sum()}")
print(f"Incorrect Date Formats Found  : {invalid_dates}")
print(f"Remaining Missing Values      : {df.isnull().sum().sum()}")
print("-" * 40)

df.to_excel("Cleaned_Data_Analytics_Dataset.xlsx", index=False)
```
